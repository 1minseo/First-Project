```python
#제거할 문자열 리스트
remove = ['ABC마트', '화장품', '악세사리', '빵집', '낚시마트', '신발마트', '주차장', '인테리어', '펫마트', '정육마트', '식당', '중고', '타이어', '로또', '열쇠','휴업중', '꽃마트','물류마트',
          '유리마트','AS', '전기차충전소', '밧데리','약국', '성인용품','침대마트', 'E-pit','수선마트','그릇','공구마트', '대기','블라인드', '가스','바이크마트','제이팜', '이 전기자재마트',
          '원예','베이커리','신발', '음식점','마트로 앙뚜아네트','비바리농','전기','타올마트','물류센터','ST','서귀주방사무용가구마트','베이킹', '&', '낚시']

#각 문자열을 순회하며 제거
for item in remove:
    df['마트명'] = df['마트명'].str.replace(item, '')

#공백 문자열 제거
df['마트명'] = df['마트명'].str.strip()

print(df['마트명'].nunique())
```
