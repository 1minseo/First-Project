```python
import json, folium
import pandas as pd
import numpy as np
from folium import Marker
from folium.plugins import MarkerCluster
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/workspace/프로젝트/최종/제주도 병원 목록.csv')
```

```python
#주소2 컬럼을 공백을 기준으로 나누기
df[['first_part', 'second_part']]=df['주소2'].str.split(n=1, expand=True)

#나눠진 데이터 중 첫번째 값을 선택하여 새로운 컬럼에 저장
df['split_addr']=df['first_part']

print(df['split_addr'])
```

```python
#데이터 개수 세기
df['split_addr'].value_counts()
```

```python
#빈도수를 백분율로 변환
addr_percentage = df['split_addr'].value_counts() / len(df['split_addr']) * 100
print(addr_percentage)
```

```python
#제주도 병원 Top5 파이차트

plt.rcParams["font.family"] = 'Malgun Gothic' #한글폰트
plt.rcParams["figure.figsize"] = (10,10) #그래픽 크기 설정

ratio = [21.73, 10.26, 8.65, 6.04, 4.43]
labels = ['노형동','서귀동','이도이동','일도이동','연동']
colors = ['#ff8e7f','#ffc000','#8fd9b6','#89a5ea','#d395d0']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(ratio, labels=labels, startangle=260,counterclock=False, colors=colors, wedgeprops=wedgeprops)

plt.legend(labels)
plt.title('제주 읍면동별 병의원 분포')
```
