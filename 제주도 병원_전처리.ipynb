```python
import pandas as pd
import numpy as np
```

```python
#데이터 불러오기
df=pd.read_excel("C:/workspace/프로젝트/병원_목록.xlsx")
df.head()
```

```python
#불필요한 컬럼 삭제
df.drop(columns=['NO', '대표전화', '월요일 진료', '화요일 진료', '수요일 진료', '목요일 진료', '금요일 진료'], inplace=True) #원본에 그대로 적용
df
```

```python
df['업무구분'].nunique()
```

```python
df['업무구분'].unique()
```

```python
#카테고리형으로 변경
df['업무구분']=df['업무구분'].astype('category')
df.dtypes
```

```python
#자료형 변경 확인
df.dtypes
```

```python
df_sorted_by_values=df.sort_values(by='업무구분', ascending=False)
df_sorted_by_values
```

```python
df[['업무','구분']]=df.업무구분.str.split(">", expand=True)
df
```

```python
df.drop(columns=['업무구분','업무'], inplace=True)
df
```

```python
df['토요일']=df['토요일 진료'].apply(lambda x:"0" if x.strip() == "" else "1")
df
```

```python
df['일요일']=df['일요일 진료'].apply(lambda x:"0" if x.strip() == "" else "1")
df
```

```python
df['공휴일']=df['공휴일 진료'].apply(lambda x:"0" if x.strip() == "" else "1")
df
```

```python
#불필요한 컬럼 삭제
df.drop(columns=['토요일 진료', '일요일 진료', '공휴일 진료'], inplace=True) #원본에 그대로 적용
df
```

```python
#분할된 결과를 새로운 df에 저장
split_addr=df['주소'].str.split(",", expand=True) #분할된 결과를 여러 열로 확장하여 반환
split_addr
```

```python
#기존 df에 새로운 열로 추가
df['주소2'] = split_addr[0]
```

```python
split_addr2=df['주소2'].str.split("(", expand=True) #분할된 결과를 여러 열로 확장하여 반환
split_addr2
```

```python
#기존 df에 새로운 열로 추가
df['주소3'] = split_addr2[0]
```

```python
df
```

```python
#불필요한 컬럼 삭제
df.drop(columns=['주소', '주소2'], inplace=True) #원본에 그대로 적용
df
```

```python
#컬럼명 변경
df.rename(columns={'주소3':'주소'}, inplace=True)
```

```python
idx = df[df['구분'].str.contains('한의원') | df['구분'].str.contains('치과의원') | df['구분'].str.contains('요양병원')
         | df['구분'].str.contains('보건소') | df['구분'].str.contains('보건지소') | df['구분'].str.contains('보건진료소') 
         | df['구분'].str.contains('한방병원') | df['구분'].str.contains('치과병원')].index
df = df.drop(idx)
```

```python
df.to_csv("제주도 병원 목록.csv", index=False)
```
